```python
# Implementing Lasso Regularization - 
# Implementing Lasso Regularization for feature engineering - 
from sklearn.linear_model import Lasso
lasso = Lasso(alpha=0.1)
lasso.fit(X_scaled,y)
```

```python
# Choosing the best alpha value is crucial -
from sklearn.linear_model import LassoCV

# Use LassoCV to perform cross-validation to find the best alpha
lasso_cv = LassoCV(alphas=[0.01, 0.1, 1, 10, 100], cv=5)
lasso_cv.fit(X_scaled, y)

# Best alpha value found
best_alpha = lasso_cv.alpha_

# Fit Lasso model with the best alpha
lasso = Lasso(alpha=best_alpha)
lasso.fit(X_scaled, y)

# Coefficients
coefficients2 = lasso.coef_
```

```python
X_scaled.shape
```

```python
selected_features2 = [i for i, coef in enumerate(coefficients2) if coef != 0]
```

```python
selected_features2 # Out of 34 we achieve 20 best features
```

```python
x = dataset3.iloc[:, selected_features2]  # Selects columns based on feature indices
y = dataset3['Target'] 
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
model = LogisticRegression(max_iter=200)
scores = cross_val_score(model, X, y, cv=5)
print(f"Cross-validated accuracy: {scores.mean():.2f}")
```

### Balancing the data - 

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

# Define your features and target
x2 =dataset3.iloc[:, selected_features2] 
y = dataset3['Target']

X_train, X_test, y_train, y_test = train_test_split(x2, y, test_size=0.2, random_state=42)

smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
X_test_smote,y_test_smote = smote.fit_resample(X_test,y_test)

# Check the new class distribution
print(y_train_smote.value_counts()) # class imbalancing is completed 
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score
clf = LogisticRegression()
clf.fit(X_train_smote, y_train_smote)  # Use the resampled data
y_pred = clf.predict(X_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
y_pred = clf.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
